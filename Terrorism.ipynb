```python
from numpy.ma.core import putmask
iraqy=data.loc[data['country_txt']=='Iraq']
iraqy=iraqy.loc[data['iyear']>=2000]

afghan=data.loc[data['country_txt']=='Afghanistan']
afghan=afghan.loc[data['iyear']>=2000]

india=data.loc[data['country_txt']=='India']
#india=india.loc[data['iyear']>=2000]

pak=data.loc[data['country_txt']=='Pakistan']
pak=pak.loc[data['iyear']>=1990]

```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 5), sharey=True)
fig.suptitle('Top Countries with Attacks over the years')

sns.histplot(ax=axes[0],data=iraqy,x='iyear',binwidth=1)
axes[0].set_title('Iraq')

sns.histplot(ax=axes[1],data=afghan,x='iyear',binwidth=1)
axes[1].set_title('Afghanistan')

sns.histplot(ax=axes[2],data=pak,x='iyear',binwidth=1)
axes[2].set_title('Pakistan')

sns.histplot(ax=axes[3],data=india,x='iyear',binwidth=1)
axes[3].set_title('India')
```
